```python
import math

class BasicCalculator:
    def add(self, x, y):
        return x + y

    def subtract(self, x, y):
        return x - y

    def multiply(self, x, y):
        return x * y

    def divide(self, x, y):
        if y == 0:
            return "Cannot divide by zero"
        return x / y

    def modulus(self, x, y):
        if y == 0:
            return "Cannot calculate modulus by zero"
        return x % y

class AdvancedCalculator(BasicCalculator):
    def logarithmic(self, x):
        if x <= 0:
            return "Invalid input"
        return math.log10(x)

    def percentage(self, x, y):
        if y == 0:
            return "Invalid input"
        return (x / y) * 100

    def power(self, x, y):
        return x ** y

    def square_root(self, x):
        if x < 0:
            return "Invalid input"
        return math.sqrt(x)

    def factorial(self, x):
        if x < 0:
            return "Factorial is not defined for negative numbers"
        if x == 0:
            return 1
        return x * self.factorial(x - 1)

def main():

    basic_calculator = BasicCalculator()
    advanced_calculator = AdvancedCalculator()

    while True:
        print("\nSelect Calculator Mode:")
        print("1. Basic Calculator")
        print("2. Advanced Calculator")

        choice = input("Enter your choice (1/2): ")

        if choice == '1':
            print("\nBasic Calculator Operations:")
            x = float(input("Enter first number: "))
            y = float(input("Enter second number: "))
            print("1. Addition")
            print("2. Subtraction")
            print("3. Multiplication")
            print("4. Division")
            print("5. Modulus")
            operation = input("Select operation (1/2/3/4/5): ")
            if operation == '1':
                print("Result:", basic_calculator.add(x, y))
            elif operation == '2':
                print("Result:", basic_calculator.subtract(x, y))
            elif operation == '3':
                print("Result:", basic_calculator.multiply(x, y))
            elif operation == '4':
                print("Result:", basic_calculator.divide(x, y))
            elif operation == '5':
                print("Result:", basic_calculator.modulus(x, y))
            else:
                print("Invalid choice!")

        elif choice == '2':
            print("\nAdvanced Calculator Operations:")
            x = float(input("Enter a number: "))
            print("1. Logarithmic")
            print("2. Percentage")
            print("3. Power")
            print("4. Square Root")
            print("5. Factorial")
            operation = input("Select operation (1/2/3/4/5): ")
            if operation == '1':
                print("Result:", advanced_calculator.logarithmic(x))
            elif operation == '2':
                y = float(input("Enter total value: "))
                print("Result:", advanced_calculator.percentage(x, y))
            elif operation == '3':
                y = float(input("Enter exponent: "))
                print("Result:", advanced_calculator.power(x, y))
            elif operation == '4':
                print("Result:", advanced_calculator.square_root(x))
            elif operation == '5':
                print("Result:", advanced_calculator.factorial(int(x)))
            else:
                print("Invalid choice!")

        else:
            print("Invalid choice")

if __name__ == "__main__":
    main()
```
